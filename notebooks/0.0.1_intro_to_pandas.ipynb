### Step 10. For the most-expensive street, what is the avg price of house?

```python
c = df.groupby('Street')
c = c.sum()
c = c.sort_values(['SalePrice'], ascending=False)
street_name = c.iloc[0].name
expensive = df[df['Street'] == street_name]
expensive['SalePrice'].mean()
```
